```python
import pandas as pd
from pandas import read_csv
from nltk.corpus import stopwords
import numpy as np
```

```python
df = read_csv('spotify_data.csv')
```

```python
# clean date 
df["track_album_release_date"] = pd.to_datetime(df["track_album_release_date"])
df['track_album_release_date'] = pd.to_datetime(df['track_album_release_date'], format='%d/%m/%y %H:%M:%S')
df.loc[df['track_album_release_date'].dt.year > 2022, 'date_clean'] = df['track_album_release_date'] - pd.DateOffset(years=100)
df.loc[df['track_album_release_date'].dt.year <= 2022, 'date_clean'] = df['track_album_release_date']
df.drop('track_album_release_date',axis=1)
```

```python
# add necessary columns
df['year'] = df['date_clean'].dt.year
```

```python
# viz_1
viz_1 = df.groupby(['year','playlist_genre'])['duration_ms'].agg('mean').to_frame(name = 'avg_duration').reset_index()
viz_1['avg_duration'] = viz_1['avg_duration'] /1000 / 60
viz_1.shape
```

```python
# viz_2
viz_2 = df.groupby(['year','playlist_genre'])['danceability','valence','energy','loudness','mode'].agg('mean').reset_index()
viz_2
```

```python
#EXTRA THINGS
df = df.astype({'track_name':'string'})
df['track_name'] = df['track_name'].str.lower()
df['track_name'] = df['track_name'].str.replace(r'[^\w\s]+', '')
df.track_name = df.track_name.fillna('')



```

```python
import nltk.corpus
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['feat','.','(',')','remix'])
stop_words
```

```python
df['track_name_clean'] = df['track_name'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```
